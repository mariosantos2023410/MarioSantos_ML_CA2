### 2.3 Dealing with Categorical Values

```python
categorical_col = [col for col in df.columns if df[col].dtype == 'object']
print('Categorical columns :',categorical_col)
```

There are six categorical columns present in this dataset as can be seen above. Therefore, these values needed to be converted to numerical values to make it suitable for machine learning algorithms to interpret. Machine Learning Algorithms cannot process objects, so these data need to be handled. For `Education` Ordinal Encoding can be used as these variables have a relationship with each other. For other variables such as `"Personal Loan", "Securities Account", "CD Account", "Online Banking", and "CreditCard"`, the Hot-Encoding as known as Dummy Variables can be used to handle them as these sub-variables do not have any relationship.


### 2.3.1 Applying Ordinal Encoding for the Variable 'Education'

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df['Education'] = encoder.fit_transform(df[['Education']])
```

The reason I have chosen this method is because according to (Fulmyk, 2023) It assigns unique numerical labels to each category, preserving the inheret ordinal relationship between categories. For example, in this situation "Diploma" may be encoded as 0, "Degree" as 1, and "Masters" as 2. Addittionaly, the Variable "Education" it is considerate to be an ordinal variable, which is more appropriate to use Ordinal Encoding.

```python
df
```

### 2.3.2 Applying Label Encoding for the rest of Categorical Variables

The reason I have chosen to apply label encoding for these variables it's because there are only two unique values such as "Yes" and "No".

```python
from sklearn.preprocessing import LabelEncoder
le_personal_loan = LabelEncoder()
df["Personal Loan"] = le_personal_loan.fit_transform(df["Personal Loan"])
```

```python
le_securities_account = LabelEncoder()
df["Securities Account"] = le_securities_account.fit_transform(df["Securities Account"])
```

```python
le_cd_account = LabelEncoder()
df["CD Account"] = le_cd_account.fit_transform(df["CD Account"])
```

```python
le_online_banking = LabelEncoder()
df["Online Banking"] = le_online_banking.fit_transform(df["Online Banking"])
```

```python
le_credit_card = LabelEncoder()
df["CreditCard"] = le_credit_card.fit_transform(df["CreditCard"])
```
