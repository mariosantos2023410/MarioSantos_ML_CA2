# 4. Model Development Setup

### 4.1 Splitting the Feature and Target Columns

```python
X = df.drop("Income(Thousands's)", axis=1)
y = df["Income(Thousands's)"]
```

### 4.2 Scaling the Data

```python
from sklearn.preprocessing import MinMaxScaler

scx= MinMaxScaler()
scy= MinMaxScaler()

X= scx.fit_transform(X)
y= y.values.reshape(-1,1)
y=scy.fit_transform(y)
```

### 4.3 Splitting Train and Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# 5. Models

### 5.1 Running Artificial Neural Network for Regression

```python
from keras import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(30, input_dim=11, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="linear"))
```

```python
model.compile(loss="mean_squared_error", optimizer="adam",  metrics=["mae"])
model.fit(X, y, epochs=100, batch_size=32)
```

```python
scores=model.evaluate(X,y)
print("%s: %.2f" % (model.metrics_names[1],scores[1]*100))
```

```python
y_pred = model.predict(X_test)
```

### 5.3 Plotting the Predict and Measured Values

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Measured Values')
ax.set_ylabel('Predicted Values')
plt.show()
```

The scatter plot above shows that the values surrounds the black broken line where it represents the predict values.

### 5.4 Reshaping the Results

```python
X_final = df.drop("Income(Thousands's)", axis=1)
y_final = df["Income(Thousands's)"]

X_final = scx.transform(X_final)
y_final = scy.transform(y_final.values.reshape(-1,1))
```

```python
results = scy.inverse_transform(model.predict(X_final))
```

### 5.5 Displaying the Predictive Results of Income

```python
print(results)
```

### 5.6 Running Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

rf = RandomForestRegressor(random_state=1)
rf.fit(X_train,y_train)

y_pred = rf.predict(X_test)

print("Mean Absolute Error (MAE):",mean_absolute_error(y_test,y_pred))
print("Root Mean Squared Error (RMSE):",np.sqrt(mean_squared_error(y_test,y_pred)))
print("R Score:",r2_score(y_test,y_pred))
```

```python
y_pred = rf.predict(X_test)
```

```python
y_pred_inverse = scy.inverse_transform(y_pred.reshape(-1, 1))
y_test_inverse = scy.inverse_transform(y_test)
```
