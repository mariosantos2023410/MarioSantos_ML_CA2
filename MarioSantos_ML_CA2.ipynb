### 2.3 Identifying Outliers

Here I am calculating the interquatile range for the Income column in the Dataset. This helps to find the outliers within the data.

```python
Q1 = np.percentile(df["Income(Thousands's)"], 25, method='midpoint')
Q3 = np.percentile(df["Income(Thousands's)"], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)
```

##### Above Upper bound

It calculates values above threshold at upper bound that are considered to be outliers.

```python
upper = Q3+1.5*IQR
upper_array = np.array(df["Income(Thousands's)"] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())
```

##### Below Lower bound

It calculates values below threshold at lower bound that are considered to be outliers.

```python
lower = Q1-1.5*IQR
lower_array = np.array(df["Income(Thousands's)"] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

In this situation here, the Inter Quartile Range it is adjusted and scaled up. The reason is to widen the range which is considered to be "normal" data, this practice is to avoid labeling too many values as outliers.
